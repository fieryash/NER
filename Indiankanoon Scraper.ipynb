```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import json
from tqdm import tqdm
# from fuzzywuzzy import fuzz
```

```python
act_url = "https://indiankanoon.org/doc/1317393"
root_url=act_url.split("doc")[0]
rr=requests.get(act_url)
rr_soup=BeautifulSoup(rr.text,'html.parser')
title=rr_soup.find("div", class_ = "doc_title").string
html_code=rr_soup.find_all("div", class_ = "article")
dictionary = {}
dictionary["Act Name"] = title
dictionary["Section"] = {}
for item in html_code:
    section_id = re.sub(r'[^\w\s]', '', item.attrs.get('id'))
    dictionary["Section"][ section_id ] = {}
    dictionary["Section"][section_id]["full_content"] = item.text
    children = item.findChildren("div", class_ = "section")
    dictionary["Section"][ section_id]["sub_section"] = {}
    for section in children:
#         print(section)
        sub_section_id = re.sub(r'[^\w\s]', '', section.attrs.get('id'))
        dictionary["Section"][ section_id]["sub_section"][sub_section_id] = {}
        dictionary["Section"][ section_id]["sub_section"][sub_section_id]['content'] = section.text
        sub_sections = section.findChildren("div", class_ = "subsection")
        dictionary["Section"][ section_id]["sub_section"][sub_section_id]["sub_sub_section"] = {}
        
        for sub_section in sub_sections:
            
            sub_sub_section_id = re.sub(r'[^\w\s]', '', sub_section.attrs.get('id'))
            dictionary["Section"][ section_id]["sub_section"][sub_section_id]["sub_sub_section"][sub_sub_section_id] = {}
            dictionary["Section"][ section_id]["sub_section"][sub_section_id]["sub_sub_section"][sub_sub_section_id]['content'] = sub_section.text
            
            sub_sub_sections = sub_section.findChildren("div", class_ = "subsubsection")
            dictionary["Section"][ section_id]["sub_section"][sub_section_id]["sub_sub_section"][sub_sub_section_id]["sub_sub_sub_section"] = {}
            for sub_sub_sub_section in sub_sub_sections:
                sub_sub_sub_section_id = re.sub(r'[^\w\s]', '', sub_sub_sub_section.attrs.get('id'))
                dictionary["Section"][ section_id]["sub_section"][sub_section_id]["sub_sub_section"][sub_sub_section_id]["sub_sub_sub_section"][sub_sub_sub_section_id] = {}
                dictionary["Section"][ section_id]["sub_section"][sub_section_id]["sub_sub_section"][sub_sub_section_id]["sub_sub_sub_section"][sub_sub_sub_section_id]['content'] = sub_sub_sub_section.text
            
                print(sub_sub_sub_section)
#             print(sub_section)
#     print(children)


```

```python
def punc(string):
    punc = '''!()-[]{};:'"\, <>./?@#$%^&*_~'''

    # Removing punctuations in string 
    # Using loop + punctuation string 
    for ele in string:  
        if ele in punc:  
            string = string.replace(ele, " ") 
            string = string.lower()
            string = " ".join(string.split())
    return string 
```
