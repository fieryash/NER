```python
def Punctuation(string): 
  
    # punctuation marks 
    punctuations = '''.'''
  
    # traverse the given string and if any punctuation 
    # marks occur replace it with null 
    for x in string.lower(): 
        if x in punctuations: 
            string = string.replace(x, "") 
  
    return string
```

```python
def clean(dictionary):
    copy = dictionary.copy()
    for keys in copy:
        count = 0
        for i in range(0,len(copy["Chapters"])):
            i = i-count
            for chapter in copy["Chapters"][i]:
                if chapter == "all_sections":
                    if dictionary["Chapters"][i][chapter] == {}:
                        dictionary["Chapters"].pop(i)
                        count += 1
    copy = dictionary.copy()
    for keys in copy:
        for i in range(0,len(copy["Chapters"])):
            for k in copy["Chapters"][i]:
                if k == "chapter_name":
                    if copy["Chapters"][i][k] == "ptericulars":
                        copy["Chapters"][i][k] = "Particulars"
                if k == "all_sections":
                    copy1 = copy["Chapters"][i][k].copy()

                    for section in copy1:
                        if "section" == "None":
                            continue
                        if "." in section:
                            new_section = Punctuation(section)
                            new_content = dictionary["Chapters"][i][k][section]
                            dictionary["Chapters"][i][k][new_section] = new_content
                            dictionary["Chapters"][i][k].pop(section)
    return dictionary
```
